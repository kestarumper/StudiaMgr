```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("k_1.csv"); df.head()
```

```python
grouped = df.groupby("N")
```

```python
grouped.mean().reset_index().plot.scatter(x="N", y="E")
```

```python
described = grouped.describe()['E']; described
```

```python
described['calculated'] = described.apply(lambda row: row['std'] / row['mean'], axis=1)
```
